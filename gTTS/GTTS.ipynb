```python
from gtts import gTTS
from IPython.display import Audio
```

```python
def speak_english(text):
    tts = gTTS(text=text, lang='en')
    tts.save("output_en.mp3")
    return Audio("output_en.mp3")

# Test
speak_english("Hello, this is a test of text to speech. How are you?")
```

```python
def speak_japanese(text):
    tts = gTTS(text=text, lang='ja')
    tts.save("output_jp.mp3")
    return Audio("output_jp.mp3")

# Test
speak_japanese("")
```

```python
import ipywidgets as widgets
from IPython.display import display, Audio, clear_output
from gtts import gTTS
import os

def create_tts_popup():
    # Widgets
    text_input = widgets.Textarea(
        value='',
        placeholder='Enter your text here...',
        layout=widgets.Layout(width='500px', height='120px')
    )
    
    language_select = widgets.Dropdown(
        options=[('English', 'en'), ('Japanese ()', 'ja')],
        value='en',
        layout=widgets.Layout(width='200px')
    )
    
    # Filename input
    filename_input = widgets.Text(
        value='my_audio',
        placeholder='Enter filename (without extension)',
        layout=widgets.Layout(width='200px')
    )
    
    submit_btn = widgets.Button(
        description='Generate Speech',
        button_style='success',
        layout=widgets.Layout(width='150px')
    )
    
    clear_btn = widgets.Button(
        description='Clear',
        button_style='warning',
        layout=widgets.Layout(width='150px')
    )
    
    save_btn = widgets.Button(
        description='Save to File',
        button_style='info',
        layout=widgets.Layout(width='150px')
    )
    
    output = widgets.Output()
    current_file = [None]
    
    def on_submit_click(b):
        with output:
            clear_output()
            text = text_input.value.strip()
            if not text:
                print(" Please enter some text!")
                return
            
            print(" Generating audio...")
            lang = language_select.value
            current_file[0] = f"output_{lang}.mp3"
            
            tts = gTTS(text=text, lang=lang)
            tts.save(current_file[0])
            
            print(f" Audio generated successfully!")
            display(Audio(current_file[0], autoplay=False))
            display(save_btn)
    
    def on_clear_click(b):
        text_input.value = ''
        with output:
            clear_output()
    
    def on_save_click(b):
        if current_file[0] and os.path.exists(current_file[0]):
            # Get filename from input
            custom_name = filename_input.value.strip()
            if not custom_name:
                custom_name = "my_audio"
            
            # Add .mp3 extension if not present
            if not custom_name.endswith('.mp3'):
                custom_name += '.mp3'
            
            # Save with custom name
            if current_file[0] != custom_name:
                os.rename(current_file[0], custom_name)
                current_file[0] = custom_name
            
            print(f" Saved as: {os.path.abspath(custom_name)}")
        else:
            print(" No audio file to save. Generate audio first!")
    
    submit_btn.on_click(on_submit_click)
    clear_btn.on_click(on_clear_click)
    save_btn.on_click(on_save_click)
    
    # Labels
    lang_label = widgets.HTML("<p style='margin:5px 0'><b>Language:</b></p>")
    file_label = widgets.HTML("<p style='margin:5px 0'><b>Filename:</b></p>")
    
    # Popup dialog
    popup = widgets.VBox([
        widgets.HTML("<h2 style='text-align:center'> Text-to-Speech Generator</h2>"),
        widgets.HTML("<p style='margin:5px 0'><b>Text:</b></p>"),
        text_input,
        widgets.HTML("<div style='height:10px'></div>"),
        widgets.HBox([
            widgets.VBox([lang_label, language_select]),
            widgets.HTML("<div style='width:20px'></div>"),
            widgets.VBox([file_label, filename_input])
        ]),
        widgets.HBox([clear_btn, submit_btn], layout=widgets.Layout(margin='15px 0')),
        output
    ], layout=widgets.Layout(
        padding='20px',
        border='2px solid #666',
        border_radius='10px',
        background_color='#f5f5f5',
        width='550px'
    ))
    
    display(popup)

# Show the popup
create_tts_popup()
```
